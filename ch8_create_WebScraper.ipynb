```python
books = soup.find_all('div', attrs = {'class': 'tombstone'})
print(len(books))
for b in books:
    print(b)
    print('_'*10)
    
```

```python
book_title = []
book_price = []
all_books_url = []
book_ISBN = []
```

```python
for b in books:
    title = b.find('p', attrs = {'class': 'tombstone-name'})
    book_title.append(title.string)
    price = b.find('div', attrs = {'class': 'price'})
    book_price.append(price.string)
```

```python
books_url = soup.find_all('a', attrs = {'class': 'btn btn-secondary'})
for u in books_url:
    all_books_url.append(u['href'])
```

```python
for u in all_books_url:
    req = Request(u, headers = headers)
    html = urlopen(req)
    soup = BeautifulSoup(html, 'lxml')
    isbn_tag = soup.find('th', text = re.compile('ISBN '))
    sib = isbn_tag.find_next(attrs = {'data-th': 'isbn'}) # can use next_sibling.next_sibling
    book_ISBN.append(sib.string)
    
```

```python
book_ISBN = [str(x).strip().replace('\n','') for x in book_ISBN]
```
